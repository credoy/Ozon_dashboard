```python
# ═══════════════════════════════════════════════════════
# ВАРИАНТ A: RAW-ссылка на GitHub (раскомментируйте и вставьте ссылку)
# ═══════════════════════════════════════════════════════
# url = 'https://raw.githubusercontent.com/ВАШ_АККАУНТ/ВАШ_РЕПО/main/Ozon_KPI_2020_2024.xlsx'
# df_raw = pd.read_excel(url, sheet_name='Исходные данные')
# df_kpi = pd.read_excel(url, sheet_name='Расчёт KPI', header=2)

# ═══════════════════════════════════════════════════════
# ВАРИАНТ B: Локальный файл (если загрузили вручную в Colab)
# ═══════════════════════════════════════════════════════
df_raw = pd.read_excel('Ozon_KPI_2020_2024.xlsx', sheet_name='Исходные данные')
df_kpi = pd.read_excel('Ozon_KPI_2020_2024.xlsx', sheet_name='Расчёт KPI', header=2)

print('✅ Данные загружены')
print('\n📋 Исходные данные:')
display(df_raw)
print('\n📊 Расчёт KPI:')
display(df_kpi.head(10))
```

```python
# Подготовка данных для визуализации
years = df_raw['Год'].values
gmv = df_raw['GMV, млрд руб.'].values
orders = df_raw['Заказы, млн шт.'].values
revenue = df_raw['Выручка, млрд руб.'].values
buyers = df_raw['Активные покупатели, млн чел.'].values
ebitda = df_raw['Adj EBITDA, млрд руб.'].values
aov = df_raw['AOV, руб.'].values
take_rate = df_raw['Take rate, %'].values

# Расчётные показатели
gmv_growth = np.round(np.diff(gmv) / gmv[:-1] * 100, 2)
rev_growth = np.round(np.diff(revenue) / revenue[:-1] * 100, 2)
orders_growth = np.round(np.diff(orders) / orders[:-1] * 100, 2)
buyers_growth = np.round(np.diff(buyers) / buyers[:-1] * 100, 2)
rev_per_buyer = np.round(revenue / buyers, 3)
orders_per_buyer = np.round(orders / buyers, 3)
gmv_per_buyer = np.round(gmv / buyers, 3)
ebitda_margin = np.round(ebitda / revenue * 100, 2)
aov_growth = np.round(np.diff(aov) / aov[:-1] * 100, 2)

print('✅ Данные подготовлены')
```

## 2. KPI-карточки (KPI Cards)

Ключевые показатели за последний отчётный год (2024) с динамикой за 5 лет.

```python
# KPI-карточки
fig_kpi = make_subplots(
    rows=1, cols=5,
    subplot_titles=[
        f'<b>GMV</b><br><span style="font-size:24px">{gmv[-1]:,.1f}</span> млрд ₽',
        f'<b>Заказы</b><br><span style="font-size:24px">{orders[-1]:,.1f}</span> млн',
        f'<b>Выручка</b><br><span style="font-size:24px">{revenue[-1]:,.1f}</span> млрд ₽',
        f'<b>Покупатели</b><br><span style="font-size:24px">{buyers[-1]:,.1f}</span> млн',
        f'<b>Adj EBITDA</b><br><span style="font-size:24px">{ebitda[-1]:,.1f}</span> млрд ₽'
    ],
    specs=[[{'type': 'indicator'}]*5]
)

# GMV
fig_kpi.add_trace(go.Indicator(
    mode="number+delta",
    value=gmv[-1],
    delta={'reference': gmv[-2], 'relative': True, 'valueformat': '.1%',
           'increasing': {'color': '#2E7D32'}, 'decreasing': {'color': '#C62828'}},
    number={'font': {'size': 1}},
    domain={'row': 0, 'column': 0}
), row=1, col=1)

# Orders
fig_kpi.add_trace(go.Indicator(
    mode="number+delta",
    value=orders[-1],
    delta={'reference': orders[-2], 'relative': True, 'valueformat': '.1%',
           'increasing': {'color': '#2E7D32'}, 'decreasing': {'color': '#C62828'}},
    number={'font': {'size': 1}},
    domain={'row': 0, 'column': 1}
), row=1, col=2)

# Revenue
fig_kpi.add_trace(go.Indicator(
    mode="number+delta",
    value=revenue[-1],
    delta={'reference': revenue[-2], 'relative': True, 'valueformat': '.1%',
           'increasing': {'color': '#2E7D32'}, 'decreasing': {'color': '#C62828'}},
    number={'font': {'size': 1}},
    domain={'row': 0, 'column': 2}
), row=1, col=3)

# Buyers
fig_kpi.add_trace(go.Indicator(
    mode="number+delta",
    value=buyers[-1],
    delta={'reference': buyers[-2], 'relative': True, 'valueformat': '.1%',
           'increasing': {'color': '#2E7D32'}, 'decreasing': {'color': '#C62828'}},
    number={'font': {'size': 1}},
    domain={'row': 0, 'column': 3}
), row=1, col=4)

# EBITDA
fig_kpi.add_trace(go.Indicator(
    mode="number+delta",
    value=ebitda[-1],
    delta={'reference': ebitda[-2], 'relative': True, 'valueformat': '.1%',
           'increasing': {'color': '#2E7D32'}, 'decreasing': {'color': '#C62828'}},
    number={'font': {'size': 1}},
    domain={'row': 0, 'column': 4}
), row=1, col=5)

fig_kpi.update_layout(
    height=200,
    margin=dict(t=60, b=20, l=20, r=20),
    showlegend=False,
    title_text='<b>KPI Ozon Holdings PLC — 2024 vs 2023</b>',
    title_x=0.5,
    title_font_size=18
)

fig_kpi.show()
```

## 3. Динамика GMV и заказов (2020–2024)

**Гипотеза H₁:** GMV и количество заказов имеют сильную положительную корреляцию.

```python
fig1 = make_subplots(specs=[[{"secondary_y": True}]])

fig1.add_trace(
    go.Bar(x=years, y=gmv, name='GMV, млрд ₽',
           marker_color='#1E88E5', opacity=0.8,
           text=[f'{v:,.1f}' for v in gmv], textposition='outside'),
    secondary_y=False
)

fig1.add_trace(
    go.Scatter(x=years, y=orders, name='Заказы, млн шт.',
               mode='lines+markers+text',
               line=dict(color='#FF6D00', width=3),
               marker=dict(size=12, symbol='diamond'),
               text=[f'{v:,.1f}' for v in orders],
               textposition='top center',
               textfont=dict(size=11)),
    secondary_y=True
)

fig1.update_layout(
    title_text='<b>Рис. 1. GMV и количество заказов Ozon (2020–2024)</b>',
    title_x=0.5, title_font_size=16,
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="center", x=0.5),
    hovermode='x unified',
    height=500
)

fig1.update_yaxes(title_text="GMV, млрд руб.", secondary_y=False,
                  tickformat=",.0f")
fig1.update_yaxes(title_text="Заказы, млн шт.", secondary_y=True,
                  tickformat=",.0f")
fig1.update_xaxes(title_text="Год", dtick=1)

# Корреляция
corr_gmv_orders = np.corrcoef(gmv, orders)[0,1]
fig1.add_annotation(
    x=2024, y=gmv[-1],
    text=f'Корреляция r = {corr_gmv_orders:.3f}',
    showarrow=False,
    font=dict(size=13, color='#2E7D32'),
    bgcolor='rgba(255,255,255,0.9)',
    bordercolor='#2E7D32', borderwidth=1,
    yshift=40
)

fig1.show()
```

## 4. Выручка и take rate

**Гипотеза H₂:** GMV и выручка коррелируют, однако take rate снижается при масштабировании.

```python
fig2 = make_subplots(specs=[[{"secondary_y": True}]])

fig2.add_trace(
    go.Bar(x=years, y=revenue, name='Выручка, млрд ₽',
           marker_color='#43A047', opacity=0.85,
           text=[f'{v:,.1f}' for v in revenue], textposition='outside'),
    secondary_y=False
)

fig2.add_trace(
    go.Scatter(x=years, y=take_rate, name='Take rate, %',
               mode='lines+markers+text',
               line=dict(color='#E53935', width=3, dash='dot'),
               marker=dict(size=12, symbol='circle'),
               text=[f'{v:.1f}%' for v in take_rate],
               textposition='top center',
               textfont=dict(size=11)),
    secondary_y=True
)

fig2.update_layout(
    title_text='<b>Рис. 2. Выручка и take rate Ozon (2020–2024)</b>',
    title_x=0.5, title_font_size=16,
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="center", x=0.5),
    hovermode='x unified',
    height=500
)

fig2.update_yaxes(title_text="Выручка, млрд руб.", secondary_y=False,
                  tickformat=",.0f")
fig2.update_yaxes(title_text="Take rate, %", secondary_y=True,
                  tickformat=".1f", range=[15, 60])
fig2.update_xaxes(title_text="Год", dtick=1)

corr_gmv_rev = np.corrcoef(gmv, revenue)[0,1]
fig2.add_annotation(
    x=2024, y=revenue[-1],
    text=f'Корреляция GMV×Выручка r = {corr_gmv_rev:.3f}',
    showarrow=False,
    font=dict(size=13, color='#2E7D32'),
    bgcolor='rgba(255,255,255,0.9)',
    bordercolor='#2E7D32', borderwidth=1,
    yshift=40
)

fig2.show()
```

## 5. Adj EBITDA и EBITDA margin

**Гипотеза H₃:** с 2023 года компания достигла операционной прибыльности при росте GMV.

```python
colors_ebitda = ['#E53935' if v < 0 else '#43A047' for v in ebitda]

fig3 = make_subplots(specs=[[{"secondary_y": True}]])

fig3.add_trace(
    go.Bar(x=years, y=ebitda, name='Adj EBITDA, млрд ₽',
           marker_color=colors_ebitda, opacity=0.85,
           text=[f'{v:,.1f}' for v in ebitda], textposition='outside'),
    secondary_y=False
)

fig3.add_trace(
    go.Scatter(x=years, y=ebitda_margin, name='EBITDA margin, %',
               mode='lines+markers+text',
               line=dict(color='#7B1FA2', width=3),
               marker=dict(size=12, symbol='square'),
               text=[f'{v:.1f}%' for v in ebitda_margin],
               textposition='top center',
               textfont=dict(size=11)),
    secondary_y=True
)

# Нулевая линия
fig3.add_hline(y=0, line_dash="dash", line_color="gray", line_width=1,
               annotation_text="Точка безубыточности",
               annotation_position="right")

fig3.update_layout(
    title_text='<b>Рис. 3. Adj EBITDA и EBITDA margin Ozon (2020–2024)</b>',
    title_x=0.5, title_font_size=16,
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="center", x=0.5),
    hovermode='x unified',
    height=500
)

fig3.update_yaxes(title_text="Adj EBITDA, млрд руб.", secondary_y=False,
                  tickformat=",.0f")
fig3.update_yaxes(title_text="EBITDA margin, %", secondary_y=True,
                  tickformat=".1f")
fig3.update_xaxes(title_text="Год", dtick=1)

# Аннотация перехода к прибыльности
fig3.add_annotation(
    x=2023, y=ebitda[3],
    text='Переход к прибыльности<br>с 2023 г.',
    showarrow=True, arrowhead=2, arrowsize=1, arrowwidth=2,
    arrowcolor='#43A047',
    ax=0, ay=-60,
    font=dict(size=12, color='#43A047'),
    bgcolor='rgba(255,255,255,0.9)',
    bordercolor='#43A047', borderwidth=1
)

fig3.show()
```

## 7. Темпы роста ключевых показателей (год к году)

```python
growth_years = years[1:]

fig5 = go.Figure()

fig5.add_trace(go.Scatter(x=growth_years, y=gmv_growth, name='GMV',
                          mode='lines+markers+text',
                          line=dict(color='#1E88E5', width=3),
                          marker=dict(size=12),
                          text=[f'{v:.0f}%' for v in gmv_growth],
                          textposition='top center'))

fig5.add_trace(go.Scatter(x=growth_years, y=rev_growth, name='Выручка',
                          mode='lines+markers+text',
                          line=dict(color='#43A047', width=3),
                          marker=dict(size=12),
                          text=[f'{v:.0f}%' for v in rev_growth],
                          textposition='bottom center'))

fig5.add_trace(go.Scatter(x=growth_years, y=orders_growth, name='Заказы',
                          mode='lines+markers+text',
                          line=dict(color='#FF6D00', width=3),
                          marker=dict(size=12),
                          text=[f'{v:.0f}%' for v in orders_growth],
                          textposition='top center'))

fig5.add_trace(go.Scatter(x=growth_years, y=buyers_growth, name='Покупатели',
                          mode='lines+markers+text',
                          line=dict(color='#00838F', width=3),
                          marker=dict(size=12),
                          text=[f'{v:.0f}%' for v in buyers_growth],
                          textposition='bottom center'))

fig5.update_layout(
    title_text='<b>Рис. 5. Темпы роста ключевых показателей (год к году)</b>',
    title_x=0.5, title_font_size=16,
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="center", x=0.5),
    xaxis_title='Год', yaxis_title='Темп роста, %',
    hovermode='x unified',
    height=500
)

fig5.update_xaxes(dtick=1)

fig5.show()
```

## 8. Динамика среднего чека (AOV)

```python
fig6 = go.Figure()

fig6.add_trace(go.Bar(x=years, y=aov, name='AOV, руб.',
                      marker_color='#5E35B1', opacity=0.8,
                      text=[f'{v:,.0f} ₽' for v in aov],
                      textposition='outside'))

# Линия тренда
z = np.polyfit(years, aov, 2)
p = np.poly1d(z)
trend_years = np.linspace(2020, 2024, 100)
fig6.add_trace(go.Scatter(x=trend_years, y=p(trend_years),
                          mode='lines', name='Тренд',
                          line=dict(color='#E53935', width=2, dash='dash')))

fig6.update_layout(
    title_text='<b>Рис. 6. Средний чек заказа (AOV) Ozon (2020–2024)</b>',
    title_x=0.5, title_font_size=16,
    xaxis_title='Год', yaxis_title='AOV, руб.',
    showlegend=True,
    height=450,
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="center", x=0.5)
)

# Аннотация снижения
fig6.add_annotation(
    x=2024, y=aov[-1],
    text=f'Снижение AOV<br>с 2020: {((aov[-1]-aov[0])/aov[0]*100):.1f}%',
    showarrow=False,
    font=dict(size=12, color='#E53935'),
    bgcolor='rgba(255,255,255,0.9)',
    bordercolor='#E53935', borderwidth=1,
    yshift=40
)

fig6.show()
```

## 9. Корреляционный анализ

```python
corr_df = pd.DataFrame({
    'GMV': gmv,
    'Заказы': orders,
    'Выручка': revenue,
    'Покупатели': buyers,
    'EBITDA': ebitda,
    'AOV': aov
})

corr_matrix = corr_df.corr().round(3)

fig7 = px.imshow(corr_matrix,
                 text_auto=True,
                 color_continuous_scale='RdYlGn',
                 aspect='auto',
                 title='<b>Рис. 7. Корреляционная матрица показателей</b>',
                 labels=dict(color='Корреляция'))

fig7.update_layout(
    title_x=0.5, title_font_size=16,
    height=500, width=600
)

fig7.show()

print('\n📊 Результаты проверки гипотез:')
print(f'   H₁: GMV ↔ Заказы:   r = {corr_matrix.loc["GMV", "Заказы"]:.3f} ✅ Подтверждена')
print(f'   H₂: GMV ↔ Выручка:  r = {corr_matrix.loc["GMV", "Выручка"]:.3f} ✅ Подтверждена')
print(f'   H₃: EBITDA > 0 с 2023: {ebitda[3]:.1f} и {ebitda[4]:.1f} млрд ₽ ✅ Подтверждена')
```

## 10. Выводы и рекомендации

### Ключевые выводы:

1. **Масштабный рост платформы:** GMV вырос с 197,4 млрд ₽ (2020) до 2 874,7 млрд ₽ (2024) — в **14,6 раза**.

2. **Подтверждение гипотез:**
   - **H₁:** GMV и заказы коррелируют с r ≈ 0,998 — рост товарооборота напрямую связан с увеличением числа заказов.
   - **H₂:** GMV и выручка коррелируют с r ≈ 0,995 — монетизация растёт пропорционально обороту.
   - **H₃:** Adj EBITDA положительна с 2023 года (6,4 → 40,1 млрд ₽) — компания достигла операционной прибыльности.

3. **Снижение take rate:** с 52,89% (2020) до 21,42% (2024) — стратегия масштабирования за счёт снижения комиссии.

4. **Снижение AOV:** с 2 671 ₽ (2020) до 1 954 ₽ (2024) — рост частоты покупок мелких товаров.

5. **Рост аудитории:** активные покупатели выросли с 13,8 млн до 56,5 млн — в **4,1 раза**.

### Рекомендации:
- Увеличить долю высокомаржинальных категорий для роста take rate.
- Развивать программы лояльности для повышения AOV.
